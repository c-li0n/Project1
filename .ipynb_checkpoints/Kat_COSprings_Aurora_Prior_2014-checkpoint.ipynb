```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ArrestbyCounty = "data/ArrestByCounty-PopByCounty.csv"
ArrestbyCountydf = pd.read_csv(ArrestbyCounty)

```

```python
ArrestbyCountydf.head()
```

```python
CountyDF= ArrestbyCountydf[["County", "Year", "Crime", "Arrests", "Age Group", "Which", "Rate" ]]
CountyDF.head()
CleanCountyDF = CountyDF.dropna(how="any")
CleanCountyDF.head()
```

```python
ColoradoSprings_dfa = CleanCountyDF.loc[(CleanCountyDF['County']=='Teller') | (CleanCountyDF['County']=='El Paso'),:]
(ColoradoSprings_dfa.head())
ColoradoSprings= ColoradoSprings_dfa.loc[(ColoradoSprings_dfa['Year']< 2014)]
(ColoradoSprings.head())
```

```python
Aurora_df = CleanCountyDF.loc[(CleanCountyDF['County']=='Adams') | (CleanCountyDF['County']=='Arapahoe'),:]
(Aurora_df.head())
Aurora= Aurora_df.loc[(Aurora_df['Year']< 2014)]
(Aurora.head())
```

```python
Aurora_group = Aurora.groupby(['Year', 'Crime']).sum()
#print(Aurora_group)
Aurora_FinalDF= Aurora_group.reset_index()
Aurora_FinalDF.head()

```

```python
ColoradoSprings_group = ColoradoSprings.groupby(['Year', 'Crime']).sum()
#print(Aurora_group)
ColoradoSprings_FinalDF= ColoradoSprings_group.reset_index()
ColoradoSprings_FinalDF.head()
```

```python
##Aggravated Assault Line graph
AuroraAA = Aurora_FinalDF[Aurora_FinalDF["Crime"]=="Aggravated Assault"]
AuroraAA.head()
COSpringsAA = ColoradoSprings_FinalDF[ColoradoSprings_FinalDF["Crime"]=="Aggravated Assault"]

plt.plot(AuroraAA.Year, AuroraAA.Arrests)
plt.plot(COSpringsAA.Year, COSpringsAA.Arrests)
first_year = 1992
last_year = 2014
x_ticks = list(range(first_year, last_year, 3))
plt.xticks(x_ticks)

plt.title("Aggravated Assault Prior to 2014")
plt.xlabel("Year")
plt.ylabel("Number of Arrests")
plt.legend(["Aurora", "Colorado Springs"],  title= "City")

plt.savefig("Aggravated Assault.png")
```
